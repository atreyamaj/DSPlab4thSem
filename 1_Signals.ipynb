### $3 \cos \left[ \frac{\pi n}{4} + \frac{\pi}{3} \right]$

```python
#write code here
n=np.arange(-20,20)
y1=3*np.cos((np.pi)*n/4+np.pi/3)
plt.stem(n,y1)
plt.show()
```

### $\frac{\sin (\pi n/6 ) }{\pi n/6}$

```python
y2=np.sin(n*np.pi/6)/(n*np.pi/6)
plt.stem(n,y2)
plt.show()
```

Is this signal periodic ? If so find the fundamental period.

```python
n2=range(-100,100)
k=np.array([0,1,2,3])
Ak=np.array([2,4,3,4])
Wk=np.array([(np.pi/20)*0.0,(np.pi/20)*2.0,(np.pi/20)*3.0,(np.pi/20)*6.0])
thetaK=[0,-np.pi/5,-np.pi/2,-np.pi/3]
sum=np.zeros((200))
for i in range(0,4):
    sum+=Ak[i]*np.cos(Wk[i]*n2+thetaK[i])
plt.plot(n2,sum)
plt.show()
```

Let  $x[n] = [2, 4,-3,1, -5, 4, 7]; -3\leq n \leq 3$.  Generate and plot the samples of the following sequences using the stem function.  (You might find fliplr function from numpy useful for computing time reversal and np.roll for shifting)

$x_1[n] = 2x[n-3] + 3 x[n+4] $<br>
$x_2[n] = x[-n-4] + x[n+4] $<br>



```python
x=np.array([0,0,0,0,2,3,-3,1,-5,4,7,0,0,0,0])
n=np.array(range(-7,8))
x1=2*np.roll(x,3)+3*np.roll(x,-4)
#x2=np.roll(np.fliplr(x),-4)
x2=np.roll(np.flipud(x),-4)+np.roll(x,-4)
f, axarr = plt.subplots(2, sharex=True)   #Sharing X-axis
axarr[0].stem(n,x1)
#axarr[0].set_title('Cosine plot')
axarr[1].stem(n,x2)
#0axarr[1].set_title('Sine plot')
plt.show()
```
