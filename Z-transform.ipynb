Q3. Compute the causal inverse of $H(z) = \frac{z^{-1}+0.5z^{-2}}{1-0.6z^{-1}+2z^{-2}}$. (use <i>scipy.signal.residuez</i>) Is the system stable?

https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.residuez.html

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
b=np.array([0,1,0.5])
a=np.array([1,-0.6,2])
r,p,k=sig.residuez(b,a)
print(r,p,k)
t1=plt.plot(p.real, p.imag, 'rx')
plt.setp(t1, markersize=12.0, markeredgewidth=2.0)
plt.show()
u_n=np.zeros(100)
u_n[50]=1
k=range(100)
n=np.array(k)
y=sig.lfilter(b,a,u_n)
print(y)
print(n)
print(u_n)
plt.plot(n,y)
plt.show()
```

Q4. Solve the system with difference equation $y[n] - 0.8y[n - 1] = 0.5(x[n] + x[n - 1])$ with the initial condition is given by $y[-1] = -2$. Plot the output sequence for input 
$~~ (a)~ x[n] = u[n] - u[n - 10] $ and  $~~ (b)~ x[n] = 2(u[n] - u[n - 10])$. Could the system be linear? Repeat the experiment
with zero initial conditions. ( Use scipy.sig.lfilter )

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
b=[0.5,0.5]
a=[1,-0.8]
y=[-2]
k=np.arange(20)
n=np.array(k)
u_n=np.zeros(20)
u_n[n>-1]=1
u1=np.zeros(20)
u1[n>9]=1
x_n=u_n-u1
print(x_n)
zi=sig.lfiltic(b,a,y)
y,t=sig.lfilter(b,a,x_n,zi=zi)
plt.stem(n,y)
plt.show()
b=[0.5,0.5]
a=[1,-0.8]
y=[0]
k=np.arange(20)
n=np.array(k)
u_n=np.zeros(20)
u_n[n>-1]=1
u1=np.zeros(20)
u1[n>9]=1
x_n=u_n-u1
zi=sig.lfiltic(b,a,y)
y,t=sig.lfilter(b,a,x_n,zi=zi)
plt.stem(n,y)
plt.show()
b=[0.5,0.5]
a=[1,-0.8]
y=[-2]
k=np.arange(20)
n=np.array(k)
u_n=np.zeros(20)
u_n[n>-1]=1
u1=np.zeros(20)
u1[n>9]=1
x_n=2*(u_n-u1)
print(x_n)
zi=sig.lfiltic(b,a,y)
y,t=sig.lfilter(b,a,x_n,zi=zi)
plt.stem(n,y)
plt.show()
b=[0.5,0.5]
a=[1,-0.8]
y=[0]
k=np.arange(20)
n=np.array(k)
u_n=np.zeros(20)
u_n[n>-1]=1
u1=np.zeros(20)
u1[n>9]=1
x_n=2*(u_n-u1)
zi=sig.lfiltic(b,a,y)
y,t=sig.lfilter(b,a,x_n,zi=zi)
plt.stem(n,y)
plt.show()
```
