Q7. Implement the following LTI system. Plot the pole-zero plot and find the magnitude and phase response.

<ol>
  $ H(z) =  \frac{1-\sqrt2 z^{-1}+0.98z^{-2}}{2-\sqrt2 z^{-1}} $
